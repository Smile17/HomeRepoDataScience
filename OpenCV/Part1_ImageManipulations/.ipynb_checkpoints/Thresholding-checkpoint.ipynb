1) Extract objects with following images using global, adaptive, otsu thresholding
2) For global thresholding show histogram before and after thresholding
3) Simulate adaptive thresholding using the following steps:

## Read data

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
import glob
imdir = 'data\\thresholding\\'
ext = ['jpg', 'png']    # Add image formats here
files = []
[files.extend(glob.glob(imdir + '*.' + e)) for e in ext]
imgs = []

for file in files:
    print(file)
    img = cv2.imread(file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imgs.append(img)
print('Count of images:', len(imgs))
```

## Utilities

```python
def plot_images(imgs, names = None, cols = 3, with_hist = False):
    n = len(imgs)
    if names is None:
        names = ["Image " + str(x) for x in range(n)]
    rows = int(np.ceil(n / cols))
    k = 2 if with_hist else 1
    fig, axs = plt.subplots(k * rows, cols, figsize=(24, 4 * k * rows))
    for ax in axs.ravel(): # remove axis from all canvas
        ax.axis("off")
    for i, img in enumerate(imgs):
        if with_hist:
            ax = axs[k * (i // cols) + 1, i % cols]
            ax.hist(img.ravel())
        ax = axs[k * (i // cols), i % cols]
        #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ax.imshow(img)
        ax.set_title(names[i])
        
# Parameters selection utility
def param_selection(image, args_dict, func):
    def nothing(x):
        pass
    # Create a window
    cv2.namedWindow('image')
    args = args_dict.keys()
    # Initialize values
    vals = np.zeros(len(args), dtype='int')
    pVals = np.zeros(len(args), dtype='int')
    for arg in args:
        val = args_dict[arg]
        cv2.createTrackbar(arg, 'image', val[0], val[1], nothing)
    for arg in args:
        val = args_dict[arg]
        cv2.setTrackbarPos(arg, 'image', val[2])

    # Get current positions of all trackbars
        for i, arg in enumerate(args):
            vals[i] = cv2.getTrackbarPos(arg, 'image')
    result = func(image, vals) # Apply changes

    while(1):
        # Get current positions of all trackbars
        for i, arg in enumerate(args):
            vals[i] = cv2.getTrackbarPos(arg, 'image')

        if np.any(np.not_equal(vals, pVals)):
            result = func(image, vals) # Apply changes
            s = ""
            for i, arg in enumerate(args):
                s = s + arg + " = " + str(vals[i]) + " "
            print(s)
            for i in range(len(args)):
                pVals[i] = vals[i]

        # Display result image
        cv2.imshow('image', result)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cv2.destroyWindow("image")
    cv2.destroyAllWindows()
    
def show_result(img):
    cv2.imshow('image',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
plot_images(imgs)
```

## Thresholding

```python
def threshold(image, thresh = 0):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    if thresh == 0:
        thresh, img = cv2.threshold(gray,thresh,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    else:
        _, img = cv2.threshold(gray,thresh,255,cv2.THRESH_BINARY)
    
    fig, axs = plt.subplots(2, 3, figsize=(24, 8))
    ax = axs[0, 0]
    ax.imshow(image)
    ax.set_title("Original")
    ax = axs[0, 1]
    ax.imshow(gray, cmap = 'gray')
    ax.set_title("Grayscaled")
    ax = axs[0, 2]
    ax.imshow(img, cmap = 'gray')
    ax.set_title("Threshold " + str(thresh))
    axs[1, 0].axis("off")
    ax = axs[1, 1]
    ax.hist(gray.ravel())
    ax.axvline(x=thresh, color = 'red')
    ax = axs[1, 2]
    ax.hist(img.ravel())
    ax.axvline(x=thresh, color = 'red')
```
